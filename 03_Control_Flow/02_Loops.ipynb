```python
# for loop starting at 0
for i in range(5):
    print(i)
```

```python
# for loop starting at index
for i in range(1, 5):
    print(i)
```

```python
# for loop incrementing 'x' values at a time
for i in range(1, 5, 2):
    print(i)
```

```python
# for loop decrementing 'x' value at a time
for i in range(10, 0, -1):
    print(i)
```

```python
# Strings
sample_string = "This is a sample variable name"
for s in sample_string:
    print(s)
```

```python
# while loop

sample = False
index = 0
while not sample:
    index += 1
    if index > 10:
        sample = True
    print(index)
```

```python
# Break statement
index = 10
while index != 0:
    print(index)
    if index < 10 :
        break
    index = index - 1
```

```python
# Continue statement
index = 0
while index < 10 :
    if index == -1 :
        continue
    else :
        print(index)
    index += 1
```

```python
# Pass statement
index = 0
while index < 10 :
    if index == 3 :
        pass
    else :
        print(index)
    index += 1
```

```python
# Nested loop
for i in range(0, 3):
    for j in range(0, 3):
        print(f"value of i : {i} and j : {j}")
```

```python
# sum of first n natural numbers
n = 10
sum = 0
for i in range(1, n + 1):
    sum += i
print(sum)
```
